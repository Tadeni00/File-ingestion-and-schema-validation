```python
# Import Libraries
import pandas as pd
import yaml
import os
```

## Step 1: Read the file using different methods

```python
# Read CSV file
df = pd.read_csv('skin_cancer.csv')
```

## Step 2: Perform basic validation on data columns

```python
# Clean df
df=df.dropna()
df=df.drop_duplicates()
df= df.iloc[:,2:-1]
df.head()
```

## Step 3: Create a YAML file and write the column names

```python
# Write the column names to a YAML file
column_names = df.columns
with open('columns.yaml', 'w') as file:
  yaml.dump(list(column_names), file)
```

## Step 4: Validate number of columns and column names with YAML

```python
# Read the YAML file
with open('columns.yaml', 'r') as file:
  yaml_columns = yaml.safe_load(file)

# Check if the number of columns match
if len(yaml_columns) == len(df.columns):
  print('Number of columns match.')
else:
  print('Number of columns do not match.')

# Check if the column names match:
if set(yaml_columns) == set(df.columns):
  print('Column names match.')
else:
  print('Column names do not match.')
```

## Step 5: Write the file in pipe-separated text file (|) in gz format.

```python
# DataFrame is named df
output_file_path = '/content/drive/MyDrive/skin_cancer.csv'
df.to_csv(output_file_path, sep='|', compression='gzip', index=False)
```

## Step 6: Create a summary of the file:

```python
# Get file size in bytes
file_size_bytes = os.path.getsize(output_file_path)

# Convert file size to megabytes
file_size_mb = file_size_bytes / (1024 ** 2)

# Get total number of rows and columns
total_rows, total_columns = df.shape

# Print the summary
print(f"Summary of the File:\n"
      f"Total Number of Rows: {total_rows}\n"
      f"Total Number of Columns: {total_columns}\n"
      f"File Size: {file_size_mb:.2f} MB")
```
